```python
# import the necessary packages
from skimage import measure
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
def mse(imageA, imageB):
    err = np.sum((imageA.astype("float") - imageB.astype("float")) ** 2)
    err /= float(imageA.shape[0] * imageA.shape[1])
    
    return err
```

```python
def compare_images(imageA, imageB, title):
    # compute mse for image
    m = mse(imageA, imageB)
    s = measure.compare_ssim(imageA, imageB)
    
    # setup figure
    fig = plt.figure(title)
    plt.suptitle("MSE: %.2f, SSIM: %.2f" % (m, s))
    
    # show first image
    ax = fig.add_subplot(1, 2, 1)
    plt.imshow(imageA, cmap = plt.cm.gray)
    plt.axis('off')
    
    # show second image
    ax = fig.add_subplot(1, 2, 2)
    plt.imshow(imageB, cmap = plt.cm.gray)
    plt.axis('off')
    
    plt.show()
```

```python
# initialize images
good_image_1 = cv2.imread("sunset-aa.jpg")
good_image_2 = cv2.imread("sunset-aa-2.jpg")

good_insta_image = cv2.imread("sunset-instagram-good.jpg")
bad_insta_image = cv2.imread("sunset-instagram-not.jpg")
```

```python
good_image_1 = cv2.resize(good_image_1, (256, 256))
good_image_2 = cv2.resize(good_image_2, (256, 256))

good_insta_image = cv2.resize(good_insta_image, (256, 256))
bad_insta_image = cv2.resize(bad_insta_image, (256, 256))
```

```python
# convert colors
good_image_1 = cv2.cvtColor(good_image_1, cv2.COLOR_BGR2GRAY)
good_image_2 = cv2.cvtColor(good_image_2, cv2.COLOR_BGR2GRAY)

good_insta_image = cv2.cvtColor(good_insta_image, cv2.COLOR_BGR2GRAY)
bad_insta_image = cv2.cvtColor(bad_insta_image, cv2.COLOR_BGR2GRAY)
```

```python
# initialize the figure
fig = plt.figure("Images")
images = ("Good Image 1", good_image_1), ("Good Image 2", good_image_2), ("Good Insta Image", good_insta_image), ("Bad Insta Image", bad_insta_image)
```

```python
for (i, (name, image)) in enumerate(images):
    print(name)
    plt.figure()
    plt.imshow(image, cmap = plt.cm.gray)
    plt.show()
```

```python
compare_images(good_image_1, good_image_2, "Image in AA vs. Image in AA")
```

```python
compare_images(good_image_1, good_insta_image, "Image in AA vs. Good Image from Insta")
```

```python
compare_images(good_image_1, bad_insta_image, "Image in AA vs. Bad Image from Insta")
```
